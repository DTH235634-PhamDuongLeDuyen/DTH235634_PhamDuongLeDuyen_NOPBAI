```python
from math import sqrt

print("Chương trình tính diện tích Tam Giác")
a=float(input("Nhập độ dài cạnh a>0: "))
b=float(input("Nhập độ dài cạnh b>0: "))     
c=float(input("Nhập độ dài cạnh c>0: "))

if a<=0 or b<=0 or c<=0 or (a+b<=c) or (a+c<=b) or (b+c<=a):
    print("Độ dài cạnh không hợp lệ")
else:
        cv=a+b+c
        p=cv/2
        s=sqrt(p*(p-a)*(p-b)*(p-c))
        print("Chu vi tam giác là:",cv)
        print("Diện tích tam giác là:",s)
```

```python
from random import randrange

while True:
    somay = randrange(1, 101)
    solandoan = 0
    win = False

    while solandoan < 7:
        solandoan += 1
        try:
            songuoi = int(input("Máy đã chọn số [1..100], mời bạn đoán: "))
        except ValueError:
            print("Vui lòng nhập một số nguyên.")
            continue

        print("Bạn đoán lần thứ", solandoan)

        if somay == songuoi:
            print("Chúc mừng bạn đoán đúng! Số máy là:", somay)
            win = True
            break
        elif somay > songuoi:
            print("Bạn đoán sai. Số máy > số bạn đoán.")
        else:
            print("Bạn đoán sai. Số máy < số bạn đoán.")

    if not win:
        print("GAME OVER! Số máy là:", somay)

    hoi = input("Bạn có muốn chơi tiếp không? (g: tiếp tục / k: kết thúc): ").lower()
    if hoi == "k":
        break

print("Cám ơn bạn đã chơi game!")
```

```python
def BMI(height, weight):
    return weight / (height ** 2)

def PhanLoai(bmi):
    if bmi < 18.5:
        return "Gầy"
    elif bmi <= 24.9:
        return "Bình thường"
    elif bmi <= 29.9:
        return "Hơi Béo"
    elif bmi <= 34.9:
        return "Béo Phì Cấp Độ 1"
    elif bmi <= 39.9:
        return "Béo Phì Cấp Độ 2"
    else:
        return "Béo Phì Cấp Độ 3"

def NguyCoBenh(bmi):
    if bmi < 18.5:
        return "Thấp"
    elif bmi <= 24.9:
        return "Trung Bình"
    elif bmi <= 34.9:
        return "Cao"
    elif bmi <= 39.9:
        return "Rất Cao"
    else:
        return "Nguy Hiểm"

height = float(input("Nhập chiều cao (m): "))
weight = float(input("Nhập cân nặng (kg): "))
bmi = BMI(height, weight)

print("BMI của bạn =", bmi)
print("Phân loại:", PhanLoai(bmi))
print("Nguy cơ bệnh:", NguyCoBenh(bmi))
```

```python
def ROI(dt, cp):
    return (dt - cp) / cp

def GoiYDauTu(roi):
    if roi >= 0.75:
        return "Nên đầu tư"
    else:
        return "Không nên đầu tư"

print("Chương trình tính ROI")
dt = int(input("Nhập Doanh Thu: "))
cp = int(input("Nhập Chi Phí: "))
roi = ROI(dt, cp)
print("Tỉ lệ ROI =", roi)
print(GoiYDauTu(roi))
```

```python
def fibonacci(n):
    if n <= 2:
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)

def listfibo(n):
    for i in range(1, n + 1):
        print(fibonacci(i), end='\t')

# Gọi hàm
print(fibonacci(9))  # In ra số Fibonacci thứ 9
listfibo(9)          # In ra dãy từ số thứ 1 đến 9
```

```python
def oscillate(start, end):
    for i in range(start, end):
        if i == 0:
            yield 0
        else:
            yield i
            yield -i

for n in oscillate(-3, 5):
    print(n, end=' ')
```
